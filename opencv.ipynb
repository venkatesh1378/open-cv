```python
import cv2
vid=cv2.VideoCapture(0)
while True:
    ret,frame=vid.read()
    #ret is boolean value
    # ret=True the frame is captured
    # ret=False the frame is not captured
    cv2.imshow('frame',frame)
    if cv2.waitKey(1)& 0xFF==ord('q'):
        break
# after the loop release the cap object
vid.release()
cv2.destroyAllWindows()
    
```

```python
import cv2
cap=cv2.VideoCapture(0)
while (True):
    #capture the frames in the video
    ret,frame=cap.read()
    #describe the type of font
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,'hello venkatesh',(50,50),font,1,(0,255,255),2,cv2.LINE_4)
    #display the results in the frame
    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
```
