```python
import cv2
video_1=cv2.VideoCapture("D:\\DEEP LEARNING\\open cv\\images\\los_angeles.mp4")
try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print("error in creating a directory")
current_frame=0
while True:
    ret,frame=video_1.read()
    if ret:
        name='./data/frame'+str(current_frame)+'.jpg'
        print("created......"+name)
        cv2.imwrite(name, frame) 
        current_frame+=1
    else:
        break
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

video_1.release()
cv2.destroyAllWindows()
```
